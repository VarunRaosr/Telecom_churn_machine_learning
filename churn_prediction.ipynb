### Telecom Dataset - taken from Kaggle competiton . Aim is to build classifier which predicts which customers are likely to churn.

#### 1. Import libraries followed by reading in the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('bigml_59c28831336c6604c800002a.csv')
X_data = dataset.drop("phone number" ,axis=1).iloc[:,1:19].values
dataset.columns
X_data[0]
```

```python
y = dataset["churn"].values
```

```python
# Next step encode our categorical variable
X_data[0]
```

## DATA Pre Processing

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
label_enc_international = LabelEncoder()
X_data[:,2]=label_enc_international.fit_transform(X_data[:,2])
label_enc_voice_mail = LabelEncoder()
X_data[:,3]=label_enc_voice_mail.fit_transform(X_data[:,3])
X_data[0]
```

```python
label_enc_output = LabelEncoder()
y = label_enc_output.fit_transform(y)
```

## split data into training and test sets

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X_data, y, test_size=0.3, random_state=42)
```

### Feature scaling on numerical columns

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow as tf
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

### Building our neural network using Tensorflow, Keras

```python
# classifier - initialize our neural n/w
classifier = Sequential()
# Adding input layer and first hidden layer
classifier.add(Dense(output_dim =10,kernel_initializer='uniform',activation='relu',input_dim= 18))
# Dense(units, activation=None, use_bias=True, kernel_initializer='glorot_uniform',
#       bias_initializer='zeros', kernel_regularizer=None, bias_regularizer=None,
#       activity_regularizer=None, kernel_constraint=None, bias_constraint=None, **kwargs)
```

```python
# Now adding the second hidden layer  so we dont have to initalize input_dim again
classifier.add(Dense(output_dim =10,kernel_initializer='uniform',activation='relu'))
```

```python
# Finally adding the output layer
classifier.add(Dense(output_dim =1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
# compile the ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

### Training the model

```python
# fitting to training set
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
# len(y_train)
```

### Testing and Evaluating the model

```python
y_pred = classifier.predict(X_test)
y_pred = y_pred>0.5
```

```python
# EVALUATION
y_test = y_test >0.5
from sklearn.metrics import classification_report,confusion_matrix
report = classification_report(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
```

# Improving our Model

```python
# Improving Our ANN Model

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.layers import Dropout
from sklearn.model_selection import GridSearchCV

def build_classifier(optimizer):
    
    classifier = Sequential()
    classifier.add(Dense(output_dim =10,kernel_initializer='uniform',activation='relu',input_dim= 18))
    classifier.add(Dropout(p=0.1))
    classifier.add(Dense(output_dim =10,kernel_initializer='uniform',activation='relu'))
    classifier.add(Dropout(p=0.1))
    classifier.add(Dense(output_dim =1,kernel_initializer='uniform',activation='sigmoid'))
    classifier.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
    
```

```python
classifier = KerasClassifier(build_classifier,batch_size = 10,epochs = 100)
accuracies = cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=5,n_jobs=1)
```

```python
print("Average Accuracy :",accuracies.mean())
accuracies.std()
```

```python
accuracies
```

```python
classifier = KerasClassifier(build_classifier,batch_size = 10,epochs = 100)
accuracies = cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=5,n_jobs=1)
```

```python
print(accuracies.mean())
accuracies.std()
```

# Tuning

```python
classifier = KerasClassifier(build_classifier)
parameters = {'batch_size':[25,30], 'epochs':[100,200],'optimizer':['adam','rmsprop']}
grid_search = GridSearchCV(classifier,param_grid=parameters,scoring='accuracy',cv = 5)

grid_search = grid_search.fit(X_train,y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
# accuracies = cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=5,n_jobs=1)
```
